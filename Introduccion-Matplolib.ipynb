## Graficar en dos ejes

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)

plt.show()
```

## Cambiar color y forma de la gráfica

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, "k--")

plt.show()
```

Podemos utilizar los colores siguientes colores:

`r` - red/rojo

`g` - green/verde

`b` - blue/azúl

`w` - white/blanco

`k` - balck/negro

`c` - cyan/azúl cyan

`m` - magenta/rosa magenta

`y` - yellow/amarillo

Los colores se pueden combinar con las siguiente figuras:

`-` - Línea recta

`--` - Línea punteada

`*` - Estrella

`+` - Símbolo más

`^` - Triángulo hacia arriba

`v` - Triángulo hacia abajo

`>` - Triángulo hacia derecha

`<` - Triángulo hacia izquierda

`.` - Punto pequeño

`o` - Punto grande

`s` - Cuadrado

`h` - Hexágono

> Nota: Los marcadores se pueden combinar con `-` y `--`. Ejemplo: Cuadrado seguido de líneas punteadas `s--`.

Puedes ver más marcadores aquí: https://matplotlib.org/api/markers_api.html

Adicionalmente podemos utilizar los colores de html/css (nombrados, hexadecimal y rgb)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, "s--", color="#84FE54")

plt.show()
```

## Gráficas con Numpy

Podemos construir nuestros ejes a partir de arreglos de numpy.

```python
import matplotlib.pyplot as plt
import numpy as np

# Genear un espacio lineal entre 0 y 10 con 21 puntos
x = np.linspace(0, 10, 21)

print(x)
```

```python
# Generar un nuevo vector `y` con los puntos del vector `x` elevados al cuadrado
y = x ** 2

print(y)
```

```python
plt.plot(x, y, 'o--', color="steelblue")
plt.show()
```

## Graficar varias series

Una serie es un par de ejes `x`, `y` (un par de arreglos), por lo que para graficar dos series, necesitamos dos pares de arreglos `x1`, `y1` y `x2`, `y2`.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-np.pi, np.pi, 21)
y1 = np.sin(x1)

x2 = np.linspace(-np.pi, np.pi, 31)
y2 = np.cos(x2)

plt.plot(x1, y1, "*-", color="red")
plt.plot(x2, y2, "s-", color="blue")

plt.show()
```
